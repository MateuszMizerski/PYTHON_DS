```python
from __future__ import division
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import matplotlib.cm as cm
import random
import math 
import mpmath as mp


def Poiseuille():
    N1 = 80 #indeks 0 tablicy odpowiada -40
    N2 = 250
    Fi = np.zeros([N2,N1])
    Fi_N = np.zeros([N2,N1])
    Fi_T = np.zeros([N2,N1])
    W = np.zeros([N2,N1])#wirowosc
    W_N = np.zeros([N2,N1])
    W_T = np.zeros([N2,N1]) 
    dz = np.float64(0.01)
    y1 = -0.4
    y2 = 0.4
    Q = -1.0
    mi = 1.0
    y = 0.0
    #obliczenia teoretyczne
    for i in range(0,N2):
        for j in range (0,N1):
            y = j*dz + y1
            Fi_T[i][j] = Q/2.0/mi*((y**3)/3.0+(y**2)/2.0*(y1+y2)+y1*y2*y)
            W_T [i][j] = Q/2/mi*(2.0*y - y1 -y2)
    
    #war brzegowe    
    for i in range (0,N2):
        y = y1
        Fi[i][0] = Q/2.0/mi*((y**3)/3.0+(y**2)/2.0*(y1+y2)+y1*y2*y)
        W[i][0] = Q/2.0/mi*(2.0*y-y1-y2)
        y = y2
        Fi[i][N1-1] = Q/2.0/mi*((y**3)/3.0+(y**2)/2.0*(y1+y2)+y1*y2*y)
        W[i][N1-1] = Q/2.0/mi*(2.0*y-y1-y2)
        
    for j in range (1,N1-1):
        y = j*dz + y1
        Fi[0][j] = Q/2.0/mi*((y**3)/3.0+(y**2)/2.0*(y1+y2)+y1*y2*y)
        W[0][j] = Q/2.0/mi*(2.0*y-y1-y2)
        Fi[N2-1][j] = Q/2.0/mi*((y**3)/3.0+(y**2)/2.0*(y1+y2)+y1*y2*y)
        W[N2-1][j] = Q/2.0/mi*(2.0*y-y1-y2)

    Fi_N = Fi
    W_N = W
    #50
    
    for p in range (0,5000):

        for i in range (1,N2-1):
            for j in range (1,N1-1):
                pom = np.float64(0.0)
                Fi_N[i][j] = (Fi[i+1][j]+Fi[i-1][j]+Fi[i][j-1]+Fi[i][j+1] - W[i][j]*(dz**2))/4.0
                pom = (W[i+1][j]+W[i-1][j]+W[i][j-1]+W[i][j+1])/4.0
                pom = pom - ((Fi[i][j+1]-Fi[i][j-1])*(W[i+1][j]-W[i-1][j])-(Fi[i+1][j]-Fi[i-1][j])*(W[i][j+1]-W[i][j-1]))/16.0
                W_N[i][j] = pom
        Fi=Fi_N
        W = W_N
        
    
    
#     while True:
        
#         for i in range (1,N2-1):
#             for j in range (1,N1-1):
#                 Fi_N[i][j] = (Fi[i+1][j]+Fi[i-1][j]+Fi[i][j-1]+Fi[i][j+1] - W[i][j]*(dz**2))/4.0
#                 W_N[i][j] = (W[i+1][j]+W[i-1][j]+W[i][j-1]+W[i][j+1])/4.0
#                 W_N[i][j] = W_N[i][j] - ((Fi[i][j+1]-Fi[i][j-1])*(W[i+1][j]-W[i-1][j])-(Fi[i+1][j]-Fi[i-1][j])*(W[i][j+1]-W[i][j-1]))/16.0
                
#         if (abs(Fi[150][40] - Fi_N[150][40])<0.0000001*abs(Fi[150][40])and abs(W[150][40] - W_N[150][40])<0.0000001*abs(W[150][40]) ):
#             break
#         #print(0.000001*abs(Fi[150][40]))
#         Fi = Fi_N
#         W = W_N
    print("END")
    return Fi,W,Fi_T,W_T


```

```python
N1 = 80 #indeks 0 tablicy odpowiada -40
N2 = 250
Fi1 = np.zeros([N2,N1])
W1 = np.zeros([N2,N1])
Fi_T = np.zeros([N2,N1])
W_T = np.zeros([N2,N1])

Fi1,W1,Fi_T,W_T = Poiseuille()
            
    
```

```python
max_W = 0.0
max_F = 0.0

for i in range (0,N2):
    for j in range(0,N1):
        if(abs(Fi_T[i][j]-Fi1[i][j])>max_F):
            max_F = abs(Fi_T[i][j]-Fi1[i][j])
        if(abs(W_T[i][j]-W1[i][j])>max_F):
            max_W = abs(W_T[i][j]-W1[i][j])
            
print("Maksymalna różnica miedzy wartością teoretyczną i doświadczalna strumienia:", max_F)
print("Maksymalna różnica miedzy wartością teoretyczną i doświadczalna wirowosci:", max_W)            
print("")
print("END")
```

```python
F_0, F_07, W_0,W_07,Y = [],[],[],[],[]
dz = 0.01
i = int (0.7/dz)
for j in range (0,N1):
    Y.append(j*dz-0.4)
    F_0.append(Fi1[0][j])
    F_07.append(Fi1[i][j])
    W_0.append(W1[0][j])
    W_07.append(W1[i][j])

plt.figure(figsize=(10.0,10.0))
plt.plot(Y,F_0,'r',label = r'$\psi$(0.0,y)')
plt.plot(Y,F_07,'g', label = r'$\psi$(0.7,y)')
plt.title(r'$\psi$(y)')
plt.xlabel(r'y')
plt.legend()
plt.xlim(-0.4,0.4)
plt.ylabel(r'$\psi$')
#plt.yscale('log')
plt.show()

```

```python
plt.figure(figsize=(10.0,10.0))
plt.plot(Y,W_0,'r',label = r'$\zeta$(0.0)')
plt.plot(Y,W_07,'g', label = r'$\zeta$(0.7)')
plt.title(r'$\zeta$(y)')
plt.xlabel(r'y')
plt.legend()
plt.xlim(-0.4,0.4)
#plt.ylim(0.0,110.0)
plt.ylabel(r'$\psi$')
#plt.yscale('log')
plt.show()
```
