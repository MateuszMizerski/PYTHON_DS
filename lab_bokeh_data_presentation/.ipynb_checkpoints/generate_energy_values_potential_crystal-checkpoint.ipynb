```python
import matplotlib.pyplot as plt
import math
import numpy as np

font = {'size'   : 22,
       'weight' : 55}
plt.rc('font', **font)


def plot_potential_chart(x: list, v: list):
    plt.figure(figsize=(10.0,10.0))
    plt.title('V(x)')
    plt.xlabel('$x[a_0]$')
    plt.ylabel('V[eV]')
    plt.ylim(-20.0,800.0)
    plt.xlim(-0.02,20.02)
    plt.plot(x,v,'black')
    plt.savefig("crystal_pot.png")
    plt.show()
    

def iteracja():
    dx = 0.025                       #krok siatki w j.a 
    N = int(20/dx)+1
    Fi = np.zeros([N])               #tablica funkcji falowej
    V = np.zeros([N])                #tab poencjalu
    #definicja potencjalu
    #*0,07357 -def bezwymiarowej
    for i in range (1,int(0.25/dx)):
        V[i] = 300*0.07537
        V[N-1-i] = 300*0.07537
    
    V[0] = 100
    V[N-1] = 100
    
    for i in range (1,9):
        V[int((0.25+2.0*i)/dx)] = 300*0.07537/2.0
        V[int((1.75+2.0*i)/dx)] = 300*0.07537/2.0
        
    for i in range (0,9):
        for j in range (1,int(0.5/dx)):
            V[int((1.75+2.0*i)/dx)+j] = 300*0.07537
    
    for i in range (0,N):
        V[i] = V[i] +(i/N*40.0-20.0)*0.07537
    
    V_R,X_R = [],[]
    for i in range(0,N):
        X_R.append(i*dx)
        V_R.append(V[i]/0.07537)

        
    energy = (x for x in np.arange(1.0, 1000.0, 0.005))
    Fi_N,E_R = [],[]
    Fi[0] = 0.0
    Fi[1] = 1.0
    
    for e in energy:
        E = e *0.07537
        xno = 0
        for i in range(1,N-1):
            Fi[i+1] = Fi[i]*(2.0+dx*dx*(V[i]-E))-Fi[i-1]
        Fi_N.append(Fi[N-1])
        E_R.append(e)
        
#         plt.plot(X_R,Fi/np.sqrt(xno))
#         plt.show()
        
    return E_R, Fi_N


def iteracja_energy_dokl(e):
    dx = 0.025                       #krok siatki w j.a 
    N = int(20/dx)+1
    Fi = np.zeros([N])               #tablica funkcji falowej
    V = np.zeros([N])                #tab poencjalu
    #definicja potencjalu
    #*0,07357 -def bezwymiarowej
    for i in range (1,int(0.25/dx)):
        V[i] = 300*0.07537
        V[N-1-i] = 300*0.07537
    
    V[0] = 100
    V[N-1] = 100
    
    for i in range (1,9):
        V[int((0.25+2.0*i)/dx)] = 300*0.07537/2.0
        V[int((1.75+2.0*i)/dx)] = 300*0.07537/2.0
        
    for i in range (0,9):
        for j in range (1,int(0.5/dx)):
            V[int((1.75+2.0*i)/dx)+j] = 300*0.07537
    for i in range (0,N):
        V[i] = V[i] +(i/N*40.0-20.0)*0.07537
    
    X, F = [],[]
        
    energy = (x for x in np.arange(e-0.005, e+0.005, 0.000001))
    Fi_N,E_R = [],[]
    Fi[0] = 0.0
    Fi[1] = 1.0
    
    X.append(0.0)
    X.append(dx)
    F.append(Fi[0])
    F.append(Fi[1])
    
    #E = e *0.07537
    xno = 0
    FN,EN = [],[]
    for en in energy:
        E = en*0.07537
        for i in range(1,N-1):
            Fi[i+1] = Fi[i]*(2.0+dx*dx*(V[i]-E))-Fi[i-1]
            #X.append((i+1)*dx)
            #F.append(Fi[i+1])
        
        FN.append(abs(Fi[N-1]))
        EN.append(en)
    ind = FN.index(min(FN))
    print(ind, EN[ind])
    X,F = iteracja_energy(EN[ind])      
    return X,F

def iteracja_energy(e):
    dx = 0.025                       #krok siatki w j.a 
    N = int(20/dx)+1
    Fi = np.zeros([N])               #tablica funkcji falowej
    V = np.zeros([N])                #tab poencjalu
    #definicja potencjalu
    #*0,07357 -def bezwymiarowej
    for i in range (1,int(0.25/dx)):
        V[i] = 300*0.07537
        V[N-1-i] = 300*0.07537
    
    V[0] = 100
    V[N-1] = 100
    
    for i in range (1,9):
        V[int((0.25+2.0*i)/dx)] = 300*0.07537/2.0
        V[int((1.75+2.0*i)/dx)] = 300*0.07537/2.0
        
    for i in range (0,9):
        for j in range (1,int(0.5/dx)):
            V[int((1.75+2.0*i)/dx)+j] = 300*0.07537
    for i in range (0,N):
        V[i] = V[i] +(i/N*40.0-20.0)*0.07537
    
    X, F = [],[]
        
    energy = (x for x in np.arange(1.0, 1000.0, 0.005))
    Fi_N,E_R = [],[]
    Fi[0] = 0.0
    Fi[1] = 1.0
    
    X.append(0.0)
    X.append(dx)
    F.append(Fi[0])
    F.append(Fi[1])
    
    E = e *0.07537
    xno = 0
    for i in range(1,N-1):
        Fi[i+1] = Fi[i]*(2.0+dx*dx*(V[i]-E))-Fi[i-1]
        X.append((i+1)*dx)
        F.append(Fi[i+1])

    return X,F
```

```python
E,F = [],[]
E,F = iteracja()
```

```python
font = {'size'   : 24,
       'weight' : 55}
plt.rc('font', **font)

plt.figure(figsize=(15.0,15.0))
plt.ylim(-1.0,1.0)
plt.xlim(0.0,1000.0)
plt.ylabel('$\Psi$')
plt.xlabel('E[ev]')
plt.plot(E,F,'black')
plt.plot([0, 1000],[0, 0], 'blue')
plt.savefig('pasma4.png')
plt.show()
```

```python
N = len(F)

l_stanu = 0
E_S,S = [],[]

#S.append(x for x in range(0, 50, 1))

for i in range(1,N):
    if(F[i]*F[i-1]<=0.0):
        S.append(l_stanu)
        if(abs(F[i])<abs(F[i-1])):
            E_S.append(E[i])
        else:
            E_S.append(E[i-1])
            
        l_stanu+=1
        if(l_stanu>=50):
            break
file = open("4.dat","w+")
ER = []
for energy in E_S:
    ER.append(str(round(float(energy),4)))
    file.write(str(round(float(energy),4)))
    file.write('\n')
file.close()
print(ER[3])
```
