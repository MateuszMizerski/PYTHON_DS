```python
import matplotlib.pyplot as plt
import math
import numpy as np

font = {'size'   : 22,
       'weight' : 55}
plt.rc('font', **font)


def wave_function_energy():
    dx = 0.025                       #krok siatki w j.a 
    N = int(20/dx)+1
    Fi = np.zeros([N])               #wave function
    V = np.zeros([N])                #potential
    #definicja potencjalu
    #*0,07357 -def bezwymiarowej
    
    V_R,X_R = [],[]
    for i in range(0,N):
        X_R.append(i*dx)
        V_R.append(V[i]/0.07537)
        
    energy = (x for x in np.arange(0.005, 1000.0, 0.005))
    Fi_N,E_R = [],[]
    Fi[0] = 0.0
    Fi[1] = 1.0
    
    for e in energy:
        E = e *0.07537
        xno = 0
        for i in range(1,N-1):
            Fi[i+1] = Fi[i]*(2.0+dx*dx*(V[i]-E))-Fi[i-1]
        Fi_N.append(Fi[N-1])
        E_R.append(e)        
    return E_R, Fi_N

def wave_function_x(e):
    dx = 0.025                       #krok siatki w j.a 
    N = int(20/dx)+1
    Fi = np.zeros([N])               #tablica funkcji falowej
    V = np.zeros([N])                #tab poencjalu
    #definicja potencjalu
    #*0,07357 -def bezwymiarowej
    
    X, F = [],[]
        
    energy = (x for x in np.arange(0.005, 1000.0, 0.005))
    Fi_N,E_R = [],[]
    Fi[0] = 0.0
    Fi[1] = 1.0
    
    X.append(0.0)
    X.append(dx)
    F.append(Fi[0])
    F.append(Fi[1])
    
    E = e *0.07537
    xno = 0
    for i in range(1,N-1):
        Fi[i+1] = Fi[i]*(2.0+dx*dx*(V[i]-E))-Fi[i-1]
        X.append((i+1)*dx)
        F.append(Fi[i+1])

    return X,F
        
```

```python
font = {'size'   : 22,
       'weight' : 55}
plt.rc('font', **font)
E,F = [],[]
E,F = wave_function_energy()
```

```python
font = {'size'   : 24,
       'weight' : 55}
plt.rc('font', **font)

plt.figure(figsize=(15.0,15.0))
plt.ylim(-1.0,1.0)
plt.xlim(0.0,1000.0)
plt.ylabel('$\Psi$')
plt.xlabel('E[ev]')
plt.plot(E,F,'black')
plt.plot([0, 1000],[0, 0], 'blue')
plt.savefig('pasma2.png')
plt.show()
```

```python
N = len(F)

l_stanu = 1
E_S,S = [],[]

#S.append(x for x in range(0, 50, 1))

for i in range(1,N):
    if(F[i]*F[i-1]<=0.0):
        S.append(l_stanu)
        if(abs(F[i])<abs(F[i-1])):
            E_S.append(E[i])
        else:
            E_S.append(E[i-1])
            
        l_stanu+=1
        if(l_stanu>=51):
            break

          
file = open("2.dat","w+")
ER = []
for energy in E_S:
    ER.append(str(round(float(energy),3)))
    file.write(str(round(float(energy),3)))
    file.write('\n')
    print(ER[-1])
file.close()
```
